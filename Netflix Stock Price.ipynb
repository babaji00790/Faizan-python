### Checking Missing Vales

```python
df.isnull().sum()
```

```python
df.duplicated().sum()

df.drop_duplicates(inplace=True)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

### create a New Column

```python
df["Daily Return"] = df["Close"].pct_change()*100
```

```python
df["Price Change"] = df["Close"] - df["Open"]
```

```python
df["MA20"] = df["Close"].rolling(20).mean()
```

```python
df["MA50"] = df["Close"].rolling(50).mean()
```

```python
df["High-Low"] = df["High"] - df["Low"]
```

```python
df["Month"] = df["Date"].dt.month_name()
```

```python
df["Year"] = df["Date"].dt.year
```

```python
df.head(10)
```

## Visualization 1

### Closing Trend Price

```python
plt.figure(figsize=(15,6))

plt.plot(df["Date"],df["Close"])

plt.title("Netflix Closing Price")

plt.xlabel("Date")

plt.ylabel("Price")

plt.show()
```

## Visualization 2

### Volume Analysis

```python
plt.figure(figsize=(15,6))

plt.plot(df["Date"],df["Volume"],color="green")

plt.title("Trading Volume")

plt.show()
```

## Visualization 3

### Moving Average

```python
plt.figure(figsize=(15,6))

plt.plot(df["Date"],df["Close"],label="Close")

plt.plot(df["Date"],df["MA20"],label="20 DMA")

plt.plot(df["Date"],df["MA50"],label="50 DMA")

plt.legend()

plt.show()
```

## Visualization 4

### Daily Return histogram

```python
plt.figure(figsize=(10,6))

sns.histplot(df["Daily Return"],bins=50)

plt.title("Daily Return Distribution")

plt.show()
```

## Visualization 5

### Correlation Heatmap

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")

plt.show()
```

## Visualization 6

### Box Plot

```python
plt.figure(figsize=(8,6))

sns.boxplot(y=df["Close"])

plt.title("Closing Price Outliers")

plt.show()
```
